```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML Lab/amazon_reviews.csv')
df.head()
```

```python
df_reviews = df.loc[:,['reviewText','overall']]
df_reviews.head()
```

```python
df_reviews.shape
```

```python
df_reviews.isnull().sum()
```

```python
df_r = df_reviews.copy()
df_r.dropna(inplace=True)
```

```python
df_r.isnull().sum()
```

```python
df_r1 = df_r.copy()
df_r1['reviewText'] = df_r1['reviewText'].str.lower()
df_r1.head()
```

```python
df_r2 = df_r1.copy()
df_r2['reviewText'] = df_r2['reviewText'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df_r2.head()
```

```python
df_r3 = df_r2.copy()
df_r3['reviewText'] = df_r3['reviewText'].apply(lambda x: x.strip())
df_r3.head()
```

```python
nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english'))

df_r4 = df_r3.copy()
df_r4['reviewText'] = df_r4['reviewText'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in stop_words]))
df_r4.head()
```

```python
df_r5 = df_r4.copy()
df_r5['reviewText'] = df_r5['reviewText'].apply(lambda x: word_tokenize(x))
df_r5.head()
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
df_r6 = df_r5.copy()
df_r6['reviewText'] = df_r6['reviewText'].apply(lambda x: [stemmer.stem(word) for word in x])
df_r6.head()
```

```python
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
df_r7 = df_r6.copy()
df_r7['reviewText'] = df_r7['reviewText'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
df_r7.head()
```
