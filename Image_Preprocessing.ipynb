```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/content/drive/MyDrive/ML Lab/Dog1.jpg')
plt.title('Original Image')
plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor (img, cv2.COLOR_BGR2GRAY)
plt.title('Gray Image')
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
crop = gray[:,300:]
plt.title('Cropped Image')
plt.imshow(crop, cmap='gray')
plt.show()
```

```python
resized = cv2.resize(crop, (256, 256))
plt.title('Resized Image')
plt.imshow(resized, cmap = 'gray')
plt.show()
```

```python
blurred = cv2.medianBlur(resized,5)
plt.title('Blurred Image (Noise Reduction)')
plt.imshow(blurred,cmap = 'gray')
plt.show()
```

```python
equ = cv2.equalizeHist(blurred)
plt.title('Contrast Enhancement')
plt.imshow(equ, cmap='gray')
plt.show()
```

```python
print(equ)
```

```python
normal = equ/255
plt.title('Normalized Image')
plt.imshow(normal, cmap='gray')
plt.show()
```

```python
print(normal)
```

```python
ret, thresh1 = cv2.threshold(normal, 0.5, 1, cv2.THRESH_BINARY)
plt.title('Binarized Image')
plt.imshow(thresh1, cmap='gray')
plt.show()
```
