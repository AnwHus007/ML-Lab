```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML Lab/indian_liver_patient.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
encoded_df = pd.get_dummies(df, columns = ['Gender'])
encoded_df.replace({True: 1, False: 0}, inplace=True)
encoded_df.head()
```

```python
corrMatrix = encoded_df.corr()
plt.figure(figsize=(12,10))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
encoded_df_1 = encoded_df.drop('Gender_Female', axis=1)
encoded_df_1.head()
```

```python
encoded_df_1.isnull().sum()
```

```python
encoded_df_2 = encoded_df_1.copy()
encoded_df_2.dropna(inplace=True)
encoded_df_2.isnull().sum()
```

```python
encoded_df_2.boxplot(figsize=(20,10))
plt.show()
```

```python
from scipy import stats
z_scores = encoded_df_2.apply(stats.zscore)
z_scores.head()
```

```python
z_scores.describe()
```

```python
outliers=  (z_scores > 2) | (z_scores < -2)
outlier_indices = outliers[outliers.any(axis=1)].index
df_no_outliers = encoded_df_2.drop(outlier_indices)
print(df_no_outliers.shape)
```

```python
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(encoded_df_2), columns=encoded_df_2.columns)
normalized_df.head()
```

```python
normalized_df.boxplot(figsize=(20,10))
plt.show()
```
